## Preguntas

¿Por año cuántos libros ha publicado un autor?
==
Recuerde que los años no trabajados deberia de aparecer en cero. considere como primer registro por autor el primer año en que publico y el ultimo el año maximo encontrado.

Ejemplo, si Uayeb public en 2009 dos libros, y en 2011 tres. y esas son las dos unicas entradas para este autor. la salida para el deberia de ser:

```python
pd.DataFrame({ "Autor" : ["Uayeb","Uayeb","Uayeb"] , "Year" : [2009,2010,2011]  , "Conteo" : [2,0,3] })
```

```python
Conteo = amazon.loc[:,["Name","Author","Year"]].groupby(["Author","Year"], as_index=False).agg({"Name":["count"]})
```

```python
Conteo.columns = Conteo.columns.droplevel(1)
```

```python
Conteo.columns = ["Author","Year","Conteo"]
```

```python
Conteo
```

¿Qué genero por año ha obtenido el mejor rating en promedio? (solo deberia de obtener un registro por año)
==

- ejemplo de la salida esperada:

```python
pd.DataFrame({
    "Year" : [2009,2010]
    , "Genre" : [ "Fiction" , "Fiction" ]
    , "Mean User Rating": [ 4.59 , 4.61 ]
})
```

```python
data3 = amazon
```

```python
data3 = amazon.loc[:,["Year","Genre","User Rating"]].groupby(["Year","Genre"], as_index=False).agg({"User Rating":["mean"]})
```

```python
data3
```

```python
data3.columns = data3.columns.droplevel(1)
```

```python
data3.columns = ["Year","Genre","Mean User Rating"]
```

```python
data3 = data3.sort_values(["Year","Mean User Rating"], ascending=True)
```

```python
data3["row_number"] = data3.groupby(["Year"], as_index = False).cumcount()
```

```python
data3
```

```python
data3 = data3.loc[data3["row_number"] == 1, :]
```

```python
data3.loc[:,["Year","Genre","Mean User Rating"]]
```

Ejemplo

Podemos ver que son dos años que este usuario ha ido decreciendo

¿Cuáles son los libros que por año se encuentran arriba del percentil 70? (Sin importar genero)
==

Ejemplo de salida esperada

```python
pd.DataFrame({
    "Year": [2010,2010,2011,2011]
    , "Titulos": ["Libro A","Libro B","Libro Z","Libro Y"]
    , "Autor": ["Autor 1","Autor 2","Autor 1","Autor 3"]
    , "Rating": [4.1, 4.0,4.2,4.3]
    , "Rating .70": [3.8,3.8,3.7,3.7]
})
```

Este ejemplo es para que se haga una idea de la estructura de salida esperada, claramente por año dependiendo de las publicaciones pueden haber varios libros arriba del percentil 70


```python
data5 = amazon
```

```python
data5 = data5.loc[:,["Year","Name","Author","User Rating"]]
```

```python
import numpy as np
```

```python
data5["Rating .70"] = np.quantile(data5["User Rating"],0.70)
```

```python
data5
```

```python
data5.loc[data5["User Rating"]>data5["Rating .70"]]
```
